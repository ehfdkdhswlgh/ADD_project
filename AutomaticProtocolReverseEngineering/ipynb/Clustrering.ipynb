```python
from scapy.all import *

int_lists = []

for packet in PcapReader('./test_capture.pcap'):

    # pcap  byte  
    byte_arr = bytes_hex(packet)
    byte_str = byte_arr.decode()

    # 2  
    hex_list = [byte_str[j:j+2] for j in range(0, len(byte_str), 2)]

    # int 
    int_list = []
    for j in range(len(hex_list)):
        int_list.append(int(hex_list[j], 16))
    int_lists.append(int_list)

# = int_lists 

result_list = int_lists
```

```python
def find_average_length(arrays): #   
    total_length = sum(len(array) for array in arrays)
    average_length = total_length / len(arrays)
    return average_length


avg_len = find_average_length(result_list)
print(avg_len)
```

```python
def adjust_array_lengths(arrays,avg_len):#       0 padding 
    #   
    for i in range(len(arrays)):
        if len(arrays[i]) >= avg_len:
            arrays[i] = arrays[i][:int(avg_len)]
        else:
            arrays[i] = arrays[i] + [0] * (int(avg_len) - len(arrays[i]))

    return arrays

result_list = adjust_array_lengths(result_list, avg_len)
print(result_list)
```
